```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from bs4 import BeautifulSoup as bs

# urllib.request for opening and reading URLs
from urllib.request import urlopen
```

```python
url = 'https://wuzzuf.net/search/jobs/?a=navbl%7Cspbl&q=data%20analyst'
```

```python
client = urlopen(url)
```

```python
html = client.read()
print(html)
```

```python
client.close()
```

```python
soup = bs(html, "html.parser")
```

```python
soup
```

```python
containers = soup.find_all("div", {"class" : "css-1gatmva e1v1l3u10"})
```

```python
len(containers)
```

```python
print(bs.prettify(containers[0]))
```

```python
containers[1].div.h2.text
```

```python
company_name = containers[0].findAll("a", {"class": "css-17s97q8"})
company_name[0].text
```

```python
f = open('wuzzuf.csv','w')
headers = "Job_Ttile, Company_Name\n"
f.write(headers)
for container in containers:
    job_title = container.find('h2', class_='css-m604qf').text.strip()
    job_company = container.find('a', class_='css-17s97q8').text.strip()
    job_location = container.find('span', class_='css-5wys0k').text.strip()
    print(job_company)
    f.write(job_title + ", " + job_company + ", " + job_location+"\n")

f.close()
```
